```python
def Capital_Gate():
    Jingmen_data = {
        '定位': '在側腰部，章門後1.8寸，當十二肋骨游離端的下方。',
        '解剖': '有腹內、外斜肌及腹橫肌；有第十一肋間動、靜脈；布有第十一肋間神經。',
        '主治': '腸鳴，泄瀉，腹脹，腰脅痛。',
        '配伍': '配行間治腰痛不可久立仰俯；配身柱、筋縮、命門治脊強脊痛。',
        '刺灸法': '斜刺0.5～0.8寸；可灸。'
    }
    return Jingmen_data

def Sun_and_Moon():
    Riyue_data = {
        '定位': '在上腹部，當乳頭直下，第7肋間隙，前正中線旁開4寸。',
        '解剖': '有肋間內、外肌，肋下緣有腹外斜肌腱膜，腹內斜肌，腹橫肌；有肋間動、靜脈；布有第七或第八肋間神經。',
        '主治': '脅肋疼痛，脹滿，嘔吐，吞酸，呃逆，黃疸。',
        '配伍': '配膽俞治膽虛；配內關、中脘治嘔吐、納呆；配期門、陽陵泉治膽石症；配支溝、丘墟治脅脹痛；配膽俞、腕骨治黃疸。',
        '刺灸法': '斜刺0.5～0.8寸；可灸。'
    }
    return Riyue_data

def Five_Shu():
    Wushu_data = {
        '定位': '在側腹部，當髂前上棘的前方，橫平臍下3寸處。',
        '解剖': '有旋髂淺深動、靜脈。分布著髂腹下神經',
        '主治': '赤白帶下，腰胯痛，少腹痛，疝氣，便秘。現多用於子宮內膜炎，睾丸炎等。',
        '配伍': '配氣海、三陰交，有調氣溫陽，散寒止痛的作用，主治少腹痛。',
        '刺灸法': '直刺0.5～1.0寸。可灸'
    }
    return Wushu_data

def Girdling_Vessel():
    Daimai_data = {
        '定位':'在側腹部，章門下1.8寸，當第12肋骨游離端下方垂線與臍水平線的交點上。',
        '解剖':'有腹內、外斜肌及腹橫肌；有第十二肋間動、靜脈；布有第十二肋間神經。',
        '主治':'月經不調，赤白帶下，疝氣，腰脅痛。',
        '配伍':'配關元、氣海、三陰交、白環俞、間使治赤白帶下；配關元、足三里、腎俞、京門、次髎治腎氣虛帶下；配中極、次髎、行間、三陰交治濕熱下注之帶下。',
        '刺灸法': '直刺0.5～0.8寸；可灸。'
    }
    return Daimai_data

def Gallbladder_Meridian_of_Foot_Shaoyang():
    acupoints_info = {
        '京門穴': Capital_Gate(),
        '日月穴': Sun_and_Moon(),
        '五樞穴': Five_Shu(),
        '帶脈穴': Girdling_Vessel()
        # ... 其他穴位
    }
    return acupoints_info

def get_acupoint_info(acupoint_name, info_category=None):
    acupoints_info = Gallbladder_Meridian_of_Foot_Shaoyang()

    acupoint_data = acupoints_info.get(acupoint_name, None)
    if acupoint_data is None:
        return "未找到指定的穴位信息。"

    if info_category:
        return acupoint_data.get(info_category, "未找到指定類別的信息。")

    return acupoint_data

def query_acupoint():
    while True:
        print("\n--- 查詢系統 ---")
        acupoint_name = input("請輸入您要查詢的穴位名稱（輸入'退出'結束查詢）:\n(京門穴、日月穴、五樞穴、帶脈穴選擇其一):")
        if acupoint_name == "退出":
            print("\n--- 已退出查詢 ---")
            print("感謝使用本系統，祝你好運～")
            break

        acupoint_info = get_acupoint_info(acupoint_name)
        if "未找到" in acupoint_info:
            print(f"\n{acupoint_info}\n")
            continue

        while True:
            print(f"\n找到穴位：{acupoint_name}")
            print("可查詢的信息類別有：定位、解剖、主治、配伍、刺灸法")
            info_category = input("請輸入您要查詢的信息類別（輸入'全部'顯示所有信息，或輸入'返回'返回上一層）:")

            if info_category == "返回":
                print("\n返回上一層...\n")
                break
            elif info_category == "全部":
                print(f"\n{acupoint_name} - 全部信息:")
                for category, detail in acupoint_info.items():
                    print(f"{category}: {detail}")
                print()  # 為了排版整潔owo
            elif info_category in ["定位", "解剖", "主治", "配伍", "刺灸法"]:
                category_info = get_acupoint_info(acupoint_name, info_category)
                print(f"\n{acupoint_name} - {info_category}:\n{category_info}\n")
            else:
                print("\n輸入的信息類別不正確，請重新輸入。\n")

            # 再次查詢或返回
            while True:
                next_action = input("輸入'再查詢'以查詢此穴位的其他信息，或輸入'返回'返回穴位選擇：")
                if next_action in ["再查詢", "返回"]:
                    break
                else:
                    print("請輸入有效的指令（'再查詢' 或 '返回'）。")
            if next_action == "返回":
                print("\n返回穴位選擇...")
                break
# 啟動查詢
query_acupoint()
```
